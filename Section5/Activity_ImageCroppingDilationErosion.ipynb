```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

image = cv2.imread('Self_Driving_Car.jpg')

cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
```

```python
height
```

```python
width
```

# Performing Cropping

```python
w0 = int(width *  0.5)
h0 = int(height * 0.5)
```

```python
h1 = int(height * 1)
w1 = int(width * 1)
```

```python
Image_cropped = image[h0:h1, w0:w1]

cv2.imshow("Crooped Image", Image_cropped)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(Image_cropped)
```

# Perform Dilation and Erosion Operations

```python
image = cv2.imread('Speed_Sign_View_2.jpg')
```

```python
kernel = np.ones((8,8), np.uint8)

image_erosion = cv2.erode(image, kernel, iterations = 1) # Try several iterations
image_dilation = cv2.dilate(image, kernel, iterations =1)

cv2.imshow('Input', image)
cv2.imshow('Erosion', image_erosion)
cv2.imshow('Dilation', image_dilation)

cv2.waitKey()
cv2.destroyAllWindows()
```
