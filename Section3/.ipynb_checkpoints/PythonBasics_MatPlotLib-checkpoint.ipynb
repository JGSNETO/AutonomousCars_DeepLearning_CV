# Matplotlib

Matplotlib is a widely-used plotting library for the Python programming language. It is particularly oseful for creating static, interactive, and animated visualizations in python.

- Library Purpose: Matplotlib is designed for creating a variety of plots and charts, including line plots, scatter plots, bar charts, histograms, and more complex visualizations. 
- Primary Audience: ItÂ´s commonly used by data scientists, statisticians, engineers, and researchers to visualize data and gain insights from it. 




## Pyplot

- Description: The 'pyplot' module is the most commonly used interface in MatPlotLib. It provides MATLAB-like way of plotting.
- Usage: Functions in 'pyplot' are stateful, which means they keep track of the current figures and plotting area.
- Plot types:
1. Line plots: Used fo connecting data points with lines
2. Scatter plots: Used for plotting individual data point
3. Bar charts: Used for categorical data representation
4. Histograms: Used for representing the distribution of data
5. Pie charts: Used for showing proportions of a whole
6. 3D plots: Used for three-dimensional data visialization

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('some numbers')
plt.show()
```

## Figure and Axes

- Figure: A 'Figure' is the top-level container for all plot elements. You can think of it as the canvas that holds all your plots. 
- Axes: An 'Axes' is an area where data is plotter; it includes thins like the X and Y axis. A single 'Figure' can contain multiple 'Axes'.

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')
ax.set_title('Title')
plt.show()
```

## Customization

- Labels and Titles: Functions to add titles and labels to the plots
- Legends: Functionality to add legends to explain what different colors/line represent
- Annotations: Adding text annotations to specific points in the plot
- Styling: Customizing colors, line styles and marker types


```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 9, 16], marker='o', linestyle='--', color='r')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Customized Plot')
plt.show()




```

## Subplot

- Multiple plots in one figure: Using 'subplots' to create multiple plots in a single figure for comparison.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax2.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.show()
```

## Interactivity

- Widgets: Matplotlib can be integrated with interactive widgets to make plots more dynamic.
- Animations: Creating animated plots to visualize changes over time. 
- ',' : Ensures the function returns a tuple

```python
from matplotlib.animation import FuncAnimation
import numpy as np

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def update(frame):
    xdata.apprend(frame)
    ydata.append(np.sin(frame))
    ln.set_data(xdata, ydata)
    return ln, 


ani = FuncAnimation(fig, update, frames =np.linspace(0, 2 * np.pi, 128), init_func=init, blit=True)

plt.show()
```
