# Convert color images to grayscale

```python
"""
## Import all the necessary libraries

- matplotlib.image (mpimg): This module is used for reading and writing image files. It provides functions to load images from files and display them using Matplotlib.
- matplotlib.pyplot (plt): This module is a state-based interface to Matplotlib, providing a MATLAB-like way of plotting. It offers functions for creating and customizing plots, such as line charts, scatter plots, histograms, and more.
- Numpy
- Opencv
"""
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Read/Display an image

image_color = mpimg.imread('image_lane_c.jpg') # Read an image and load as a numpy array. Each element of the array represents a pixel in the image, with value corresponding to RGB.
plt.imshow(image_color)


```

```python
image_color.shape # Retrieves the shape of the image
```

```python
# Convert an image into grayscale

image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray') # cmap stands for colormap, which is the parameter that specifies the color scheme used to map the pixel values to colors. 


```

```python
image_gray.shape
```

```python
image_gray
```

```python
np.set_printoptions(threshold=np.inf)# This function configures how NumPy arrays are printed. threshold=np.inf: This argument sets the threshold for summarizing array output. Using np.inf (infinity) ensures that all elements of the array are printed, and no summarization or truncation occurs.
image_gray
```

```python
# Save the grayscale image
cv2.imwrite("image_lane_g.jpg", image_gray)
```
