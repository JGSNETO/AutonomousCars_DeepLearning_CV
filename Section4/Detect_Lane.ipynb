# Lane detection through color selection

```python
# Importing all libraries
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

# Read the colored image and convert to gray sclale

image_color = mpimg.imread('image_lane_c.jpg')
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap =  'gray')
image_gray.shape
```

```python
# Lets sort out the white color out of the image
# Pixels that are white, they are in 255 range.

lane_detection_image = np.copy(image_gray)
lane_detection_image.shape
```

```python
# boolean index or masking
lane_detection_image[(lane_detection_image[:,:] < 250)] = 0 # Any value that is not white color
"""
- lane_detection_image[:, :]: This selects all elements of the array. 
The notation [:, :] means "all rows and all columns," which is essentially the entire image.
- lane_detection_image[:, :] < 250: This creates a boolean mask, a 2D array of the same shape as lane_detection_image,
where each element is True if the corresponding element in lane_detection_image is less than 250, and False otherwise.
- lane_detection_image[(lane_detection_image[:, :] < 250)]: This uses the boolean mask to select all elements in lane_detection_image that are less than 250.
= 0: This assigns the value 0 to all selected elements. 
"""

plt.imshow(lane_detection_image, cmap = 'gray') # Display the grayscale image.
plt.show() # Render so you can see in a window or inline in a notebook. 
```
