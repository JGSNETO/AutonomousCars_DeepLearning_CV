# RGB2HSV

```python
image = cv2.imread('Self_Driving_Car.jpg')
cv2.imshow('Self_Driving_Car', image)

hsvImg = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV IMAGE', hsvImg)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(hsvImg[:, :, 0])
plt.title('Hue channel')
```

```python
plt.imshow(hsvImg[:, :, 1])
plt.title('Saturation channel')
```

# Split B, G, R Channels

```python
image = cv2.imread('Self_Driving_Car.jpg')
B, G, R = cv2.split(image)
B.shape
```

```python
R.shape
```

```python
# View different channels
cv2.imshow("Blue channel", B)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# lets try to create our own 3D image out of the blue channel
zeros = np.zeros(image.shape[:2], dtype ="uint8")

cv2.imshow("Blue channel!", cv2.merge([B, zeros, zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# let's merge RGB channels to create our original image
image_merged = cv2.merge([B,G,R])
cv2.imshow("Merged image", image_merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Let's merge our original image but adding more green
image_merged = cv2.merge([B, G+100, R])
cv2.imshow("Merged image + green", image_merged)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
